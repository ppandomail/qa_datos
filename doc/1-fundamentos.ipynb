# Fundamentos de la calidad de datos
---

## Concepto de calidad de datos

* Es la medida en que los datos son aptos para su uso en operaciones, toma de decisiones y planificación, garantizando que sean precisos, completos y fiables
* Es una combinación de **personas, procesos y tecnología** enfocados a mejorar la calidad de los datos existentes en la organización
    * las **personas** definen las reglas de negocio, gestionan los datos, interpretan los resultados y toman decisiones basadas en los datos
    * los **procesos** son las políticas y procedimientos que se implementan para mantener y mejorar la calidad de los datos. Se definen estándares, implementación de prácticas como la validación y el monitoreo continuo
    * la **tecnología**, permite automatizar y optimizar los procesos relacionados con la calidad de datos. Limpieza y transformación de los datos hasta la identificación de anomalías y la evaluación de reglas de calidad

## Dimensiones de calidad de datos

Las dimensiones son métricas utilizadas para evaluar si los datos cumplen con los requisitos de negocio: 

| Dimensiones                   |                                                                                                                            |
| --                            | --                                                                                                                         |
| **Precisión (Exactitud)**     | indica cuánto reflejan los datos la realidad objetiva o una fuente verificable                                             |
| **Integridad (Completitud)**  | asegura que no falten datos necesarios; todos los registros requeridos están presentes                                     |
| **Consistencia (Coherencia)** | garantiza que los mismos datos sean uniformes entre diferentes sistemas, tablas o copias, sin contradicciones              |
| **Validez**                   | verifica que los datos se ajusten a formatos, tipos, rangos y reglas de negocio definidas (ej. fecha de nacimiento válida) |
| **Unicidad (Deduplicación)**  | asegura que cada registro aparezca una sola vez, sin duplicaciones                                                         |
| **Puntualidad (Actualidad)**  | mide si la información está disponible cuando se necesita y si está lo suficientemente actualizada                         |
| **Disponibilidad/A11y**       | se refiere a la facilidad con la que los usuarios autorizados pueden acceder a los datos cuando los necesitan              | 

## Ciencia de datos

* Es un campo interdisciplinar que haciendo uso del método científico extrae conocimiento de un conjunto de datos disponible
* Se enfoca en el estudio y análisis para obtener información significativa, conocimiento y dominio del tema para transformar los datos crudos en información valiosa
* Es el proceso de transformar datos crudos en decisiones accionables mediante el uso de análisis estadístico y computacional
* Descubrir el "porque" detrás de los datos

## Científico de datos

* Es el arquitecto de la inteligencia de datos, quien se sumerge profundamente en los conjuntos de datos, desarrolla modelos avanzados y utiliza algoritmos de aprendizaje automático para predecir tendencias y patrones futuros

## Tecnologías

* **Lenguajes de programación**: Python, R
* **Bibliotecas**: NumPy, Pandas, Matplotlib, Seaborn o Plotly, Scikit-Learn y TensorFlow
* **Bases de datos**: SQL, base de datos relacional. NoSQL, como MongoDB o Cassandra
* **Herramientas de Big Data**: Hadoop y Apache Spark
* **Cloud Computing**: Amazon Web Services (AWS), Google Cloud Platform (GCP) y Microsoft Azure para alojar y procesar datos.
* **Herramientas de visualización de datos**: Tableau, Power BI, Kaggle, Looker

## Ciclo de vida de los datos

1. **Generación y captura**:
    * Captura Manual (formularios u hojas de cálculo)
    * Sensores
    * Web Scraping: extraer datos de páginas web
    * Integración de Sistemas: (APIs, ETL)
    * Captura Automatizada de Registos. Registros de red o tráfico de dato
1. **Almacenamiento**: en sistemas o bases de datos
    * Datos no estructurados (Imágenes, textos, PDF)
    * Datos estructurados (BBDD Relacionales, CSV)
    * Datos semi estructurados (XML, JSON)
1. **Tratamiento**:
    * **Limpieza de Datos**: se identifican y corrigen los errores, valores atípicos y datos faltantes. Eliminar duplicados
    * **Transformación de Datos**: normalización de datos, conversión de unidades de medida, la codificación de variables categóricas y la agregación de datos en diferentes niveles de granularidad
    * **Control de Calidad de Datos**: se establecen criterios y métricas para evaluar: integridad, consistencia y la exactitud
    * **Segmentación de Datos**: se crean subconjuntos para un análisis más detallado
    * **Gestión de Cambios**: establecer procesos para gestionar cambios en los datos y garantizar que la información se mantenga actualizada
1. **Análisis**:
    * **Exploración de Datos**: generación de gráficos, tablas y resúmenes estadísticos para poder comprender mejor la distribución de los datos y evaluar la correlación entre variables
    * **Análisis Estadístico**: para comprender la variabilidad de los datos y realizar inferencias sobre poblaciones o fenómenos. Se puede realizar pruebas de hipótesis, análisis de varianza o regresiones para obtener información significativa
    * **Aprendizaje Automático (ML)**: se utilizan algoritmos de aprendizaje automático para crear modelos predictivos y descriptivos. Esto incluye clasificación, regresión, agrupación y tareas de procesamiento del lenguaje natural (NLP). Las herramientas populares incluyen scikit-learn, TensorFlow, y PyTorch
1. **Visualización**: comunicar de manera efectiva la información a los interesados y tomar decisiones informadas
    * **Selección de Tipos de Gráficos**: gráficos de barras, circulares, de dispersión hasta mapas de calor o líneas de tiempo. Esta elección depende de la naturaleza de los datos
    * **Diseño de Gráficos Efectivos**: deben ser claros, legibles y atractivos. Se deben elegir colores, etiquetas y títulos de manera cuidadosa para una compresión fácil y rápida
    * **Tableros de Control (Dashboards)**: conjunto de visualizaciones que se utilizan para mostrar un panorama completo de los datos y las métricas. Son útiles para monitoreo a tiempo real o toma de decisiones empresariales
    * **Herramientas**: Tableau, Power BI, Matplotlib, Seaborn, D3.js

## Gobierno de datos

* Establece las políticas, roles, responsabilidades y procedimientos necesarios para asegurar que los datos sean gestionados de manera eficiente en la organización
* Por ejemplo, define como se deben clasificar y proteger los datos sensibles y personales, y quién tiene la autoridad para modificar o acceder a ellos dentro de la organización

## Gestión de datos

* Se enfoca en las operaciones cotidianas y en la administración técnica de los datos
* Implementar una arquitectura de datos que organice eficientemente la información y garantice su seguridad y disponibilidad, son por ejemplo prácticas propias de la Gestión de Datos
* El proposito de la calidad de datos dentro de un marco de Gobierno y Gestión de datos, es evaluar y mejorar los datos para asegurar que sean aptos para los fines a los que están destinados, particularmente para la toma de decisiones y los análisis.
