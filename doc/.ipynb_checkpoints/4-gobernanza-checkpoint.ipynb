# Gobernanza

## Gobernanza de datos y calidad

## Políticas, procedimientos y roles

### Políticas de datos

Son las directrices de alto nivel que establecen los estándares de comportamiento organizacional respecto a la información
* **Seguridad y Privacidad**: define niveles de acceso y cumplimiento normativo como el RGPD
* **Calidad de Datos**: estándares de exactitud, integridad y oportunidad que deben cumplir los activos de información
* **Retención y Eliminación**: reglas sobre cuánto tiempo se guardan los datos y cómo se borran de forma segura
* **Uso Ético**: parámetros sobre el uso responsable de los datos para generar confianza

### Procedimientos y Procesos

Son las "instrucciones" detalladas para ejecutar las políticas en el día a día
* **Gestión de Metadatos**: procesos para catalogar el significado, origen y linaje de los datos
* **Resolución de Problemas**: protocolos para identificar y corregir errores de calidad detectados en las tablas operativas
* **Control de Acceso**: pasos técnicos para solicitar, aprobar y auditar quién ve qué información
* **Monitoreo**: uso de métricas para verificar que las políticas se estén cumpliendo realmente

### Roles y Responsabilidades

El éxito del gobierno depende de asignar a las personas adecuadas en una estructura
jerárquica

| Rol                          | Responsabilidad Principal                                                                             | Perfil                  |
| --                           | --                                                                                                    | --                      |
| **Chief Data Officer (CDO)** | Máximo líder de la estrategia. Alinea los datos con los objetivos del negocio                         | Ejecutivo / Estratégico |
| **Data Owner**               | Responsable final de un dominio de datos (ej. Finanzas). Autoriza el acceso y define su uso           | Líder de Negocio        |
| **Data Steward**             | Vela por la calidad e integridad operativa. Es el puente entre el negocio y la tecnología             | Experto en Datos        |
| **Data Custodian**           | Se encarga de la parte técnica: almacenamiento, copias de seguridad y seguridad de la infraestructura | TI / Arquitectura       |
