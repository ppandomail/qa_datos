## Correlación

Grado de relación entre dos variables, en el que se busca determinar qué tan bien una ecuación lineal (UNA RECTA), describe o explica la relación entre dos variables

**Coeficiente de Correlación de Pearson (r)**
* Va de 1 a -1, siendo 0 la ausencia de correlación

## Ejemplo

* Se quiere predecir la nota en base a las horas de estudio
* Como la variable independiente predice la variable dependiente
* La variable independiente tiene que estar correlacionada con la dependiente

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'hs_estudio': [3, 6, 8, 2, 1, 7],
    'nota': [8, 10, 15, 7, 5, 12]
}

df = pd.DataFrame(datos)
df
```

* **x=hs_estudio**: variable independiente
* **y=nota**: variable dependiente

### Diagrama de dispersión

```python
df.plot(kind = 'scatter', x = 'hs_estudio', y = 'nota')
plt.show()
```

### Ecuación de la recta:   $y = b_0 + b_1x$ y $R^2$

```python
x = df['hs_estudio']
y = df['nota']

x_mean = x.mean()
y_mean = y.mean()

# Pendiente e intercepto
numerador = sum((x - x_mean) * (y - y_mean))
denominador = sum((x - x_mean) ** 2)
b1 = numerador / denominador            # pendiente
b0 = y_mean - b1 * x_mean        # intercepto

# Valores predichos
y_pred = b1 * x + b0

# Calcular R^2
ss_total = sum((y - y_mean) ** 2)
ss_res = sum((y - y_pred) ** 2)
r2 = 1 - (ss_res / ss_total)

print(f'y = {b0} + {b1}x')
print(f'R^2 = {r2}')
```

* **Ecuación de la recta** es la que logra disminuir la distancia entre los puntos y la recta (Relación de mínimos cuadrados)
* **R^2**:
    * mide qué tan bien el modelo lineal explica la variabilidad de los datos (confiabilidad de la línea de tendencia)
    * cuanto más cerca esté a 1, mejor se ajustará la línea de tendencia a los datos
    * en este caso existe una buena correlación entre las variables

```python
data = []

for i in range(11):
    data.append([i, b0, b1, b0 + b1*i])

df_f = pd.DataFrame(data, columns=['hs_estudio (x)', 'intercepto (b0)', ' pendiente (b1)', 'nota (y)'])
df_f
```

**Inferir: cuantas horas hay que estudiar para alcanzar la nota máxima:**

```python
(100 - b0) / b1
```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersión
plt.scatter(x, y, color='blue', label='Datos reales')

# Recta de regresión
plt.plot(x, y_pred, color='red', label='Recta de regresión')

# Texto con la ecuación y R²
equation_text = f'y = {b0:.2f} + {b1:.2f}x\nR² = {r2:.3f}'
plt.text(float(x.min()), float(y.max()), equation_text, fontsize=12, color='black',
         bbox=dict(facecolor='white', alpha=0.7))

plt.xlabel('Horas de estudio')
plt.ylabel('Nota')
plt.title('Regresión lineal simple')
plt.legend()
plt.show()
```

### Ecuación de la recta (usando scikit-learn):   $y = b_0 + b_1x$ y $R^2$

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

datos = {
    'hs_estudio': [3, 6, 8, 2, 1, 7],
    'nota': [8, 10, 15, 7, 5, 12]
}

df = pd.DataFrame(datos)

x = df[['hs_estudio']]   # scikit-learn espera una matriz (2D)
y = df['nota']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(x, y)

# Obtener pendiente e intercepto
b1 = modelo.coef_[0]
b0 = modelo.intercept_

# Calcular R^2
r2 = modelo.score(x, y)

print(f'y = {b0} + {b1}x')
print(f'R^2 = {r2}')
```

```python
import matplotlib.pyplot as plt

# Predicciones
y_pred = modelo.predict(x)

# Gráfico
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', label=f'Regresión lineal\nR²={r2:.3f}')

plt.xlabel('Horas de estudio')
plt.ylabel('Nota')
plt.title('Regresión lineal simple')
plt.legend()
plt.show()
```
