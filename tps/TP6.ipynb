# Los datos como decisión: calidad, impacto y responsabilidad
---

## Objetivo

Integrar los conceptos del curso y reflexionar sobre el rol ético, técnico y organizacional en la gestión de datos

## Pregunta integradora

Si una organización toma decisiones críticas basadas en datos incompletos, inconsistentes o mal gobernados, ¿quién es responsable del impacto de esas decisiones y qué debería haber cambiado en el proceso de gestión de datos para evitarlo?

## Instrucciones

1. Respondé en forma breve y argumentada (máximo 15-20 líneas).
2. Podés incorporar conceptos de cualquiera de los módulos vistos (calidad, profiling, métricas, gobernanza, roles, etc.).
3. Se valorará la capacidad de conectar teoría con reflexión crítica sobre el impacto organizacional.

## Reflexión

La pregunta apunta al corazón de la **responsabilidad en la gestión de datos**. Cuando una organización toma decisiones críticas basadas en datos incompletos, inconsistentes o mal gobernados, la **responsabilidad es compartida**:
* **Alta dirección y responsables de negocio**: porque deciden sin exigir garantías de calidad ni procesos de validación.
* **Gobernanza de datos (Data Owners, Data Stewards, Comité de Datos)**: porque su rol es definir estándares, monitorear indicadores y asegurar que los datos sean confiables antes de usarse.
* **Equipos técnicos (ingeniería, analistas, científicos de datos)**: porque deben implementar controles, documentar procesos y alertar sobre riesgos de calidad.

En síntesis, no hay un único culpable: la responsabilidad es **transversal** y refleja un bajo **nivel de madurez en gestión de datos**: ausencia de gobernanza, métricas de calidad (completitud, consistencia, unicidad, validez), y cultura organizacional centrada en datos

**¿Qué debería haber cambiado?**
* **Gobernanza clara**: políticas, roles y responsabilidades definidos para custodiar los datos.
* **Ciclo de calidad continuo**: perfilado, limpieza, validación y monitoreo con indicadores (completitud, consistencia, exactitud, unicidad).
* **Transparencia y trazabilidad**: registrar fuentes, transformaciones y reglas aplicadas.
* **Alertas y dashboards ejecutivos**: que permitan detectar problemas antes de que impacten en decisiones críticas.
* **Cultura organizacional centrada en datos**: formación y concientización para que todos comprendan el valor y los riesgos de trabajar con información deficiente.

**En conclusión**: la **responsabilidad es colectiva**, y lo que debe cambiar es el **proceso de gestión de datos en su conjunto**, desde la gobernanza hasta la cultura organizacional. Es decir, construir una gestión madura, transveersal y ética que asegure que las decisiones críticas se basen en información confiable.
